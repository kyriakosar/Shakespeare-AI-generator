```python
import requests
import random
import re
import pronouncing as pr
import syllapy
```

```python


def preprocess_verse(verse):
    """Lowercase and filter all punctuation"""
    # Lowercase
    verse_lower = verse.strip().lower()
    # Filter punctuation (keep numbers)
    return re.sub('[^a-z \d]+', '', verse_lower)

def last_word(verse):
    """Retrieve the last word of a verse"""
    return preprocess_verse(verse).split()[-1]

def verse_rhyme(verse):
    """Gets the rhyme of a verse"""
    # Get the last word
    w = last_word(verse)
    # Get the last two phones
    phones = pr.phones_for_word(w)
    if phones:
        last = re.match('.+ ([a-zA-Z]+[0-9][^0-9]*)$', phones[0])
        if last:
            rhyme = last.group(1)
            return rhyme
        else:
            return phones[0]
    else:
        return ""

def countWordSyllables(word):
    return syllapy.count(word)

#     return syllable_count
def count_syllables(phrase):
    total_syllables = 0
    words = phrase.split(' ')
    for word in words:
        total_syllables += countWordSyllables(word)
    return total_syllables

# #Example usage with a poem
poem = [
"If hairs be wires black wires grow on",
"Speak of my speaking breast where-through the sun",
"the beast that bears me tired with my life on",
"gives nothing but when yellow leaves or none",
"For where is she so fair whose unear'd womb",
"When in dead night thy fair health recounting",
"to follow that which governs me to whom",
"do show thee how to make me give warning",
"That it could so preposterously be",
"And purest faith unhappily forsworn",
"therefore mayst without thy help by me be",
"to thee which thou must leave ere you were born",
"Against thy reasons making no defence",
"shows you when you look what is your substance"
]


# Initialize variables to count lines with 10 syllables
lines_with_10_syllables = 0
total_lines = len(poem)

# Extract and assign unique rhyme patterns to letters
rhyme_pattern_dict = {}
assigned_letters = []
current_letter = 'A'

for line in poem:
    # print(verse_rhyme(line))  # Analyze rhyme
    syllable_count = count_syllables(line)  # Count syllables
    if syllable_count == 10:
        lines_with_10_syllables += 1
    rhyme = verse_rhyme(line)
    if rhyme not in rhyme_pattern_dict:
        rhyme_pattern_dict[rhyme] = current_letter
        current_letter = chr(ord(current_letter) + 1)

    assigned_letters.append(rhyme_pattern_dict[rhyme])

# Create the rhyme scheme in ABAB CDCD EFEF GG format
rhyme_scheme = ''.join(assigned_letters)


print("Rhyme Scheme:", rhyme_scheme)


def count_matching_patterns(rhyme_scheme):
    matching_count = 0
    total_positions = len(rhyme_scheme)
    
    if rhyme_scheme[0] == rhyme_scheme[2]:
        matching_count += 1
    if rhyme_scheme[1] == rhyme_scheme[3]:
        matching_count += 1
    if rhyme_scheme[4] == rhyme_scheme[6]:
        matching_count += 1
    if rhyme_scheme[5] == rhyme_scheme[7]:
        matching_count += 1
    if rhyme_scheme[8] == rhyme_scheme[10]:
        matching_count += 1
    if rhyme_scheme[9] == rhyme_scheme[11]:
        matching_count += 1
    if rhyme_scheme[12] == rhyme_scheme[13]:
        matching_count += 1

    
    matching_percentage = (matching_count / (total_positions // 2)) * 100  # Divide by 2 as we only check every other position
    
    return matching_percentage

match_percentage = count_matching_patterns(rhyme_scheme)
match_percentage = round(match_percentage, 2)
print(f"{match_percentage}% of lines written with the expected rhyme scheme")


# Calculate the percentage of lines with 10 syllables
percentage_with_10_syllables = (lines_with_10_syllables / total_lines) * 100
percentage_with_10_syllables = round(percentage_with_10_syllables, 2)
print(f"{percentage_with_10_syllables}% of lines written with iambic pentameter")
```
